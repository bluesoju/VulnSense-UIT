```python
from torch_geometric.loader import DataLoader

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

for step, data in enumerate(train_loader):
    print(f'Step {step + 1}:')
    print('=======')
    print(f'Number of graphs in the current batch: {data.num_graphs}')
    print(data)
    print()
```

```python
from torch.nn import Linear
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.nn import global_mean_pool


class GCN(torch.nn.Module):
    def __init__(self, hidden_channels):
        super(GCN, self).__init__()
        torch.manual_seed(12345)
        self.conv1 = GCNConv(1536, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, hidden_channels)
        self.conv3 = GCNConv(hidden_channels, hidden_channels)
        # self.conv4 = GCNConv(128, 64)
        self.lin = Linear(hidden_channels, 3)

    def forward(self, x, edge_index, batch):
        # 1. Obtain node embeddings
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = self.conv2(x, edge_index)
        x = x.relu()
        x = self.conv3(x, edge_index)
        # x = x.relu()
        # x = self.conv4(x, edge_index)

        # 2. Readout layer
        x = global_mean_pool(x, batch)  # [batch_size, hidden_channels]

        # 3. Apply a final classifier
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.lin(x)

        return x

model = GCN(hidden_channels=32)
print(model)
```

```python
from IPython.display import Javascript
import numpy as np
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import time


# display(Javascript('''google.colab.output.setIframeHeight(0, true, {maxHeight: 300})'''))

model = GCN(hidden_channels=32)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.CrossEntropyLoss()

def train(epoch):
    model.train()
    cc = np.zeros((2, 3))
    for data in test_loader:  # Iterate in batches over the training dataset.
         out = model(data.x, data.edge_index, data.batch)  # Perform a single forward pass.
         loss = criterion(out, data.y)  # Compute the loss.
         loss.backward()  # Derive gradients.
         optimizer.step()  # Update parameters based on gradients.
         optimizer.zero_grad()  # Clear gradients.
         if epoch == 1:
          cc = np.concatenate((cc, out.detach().numpy()), axis=0)
    if epoch == 1:
      cc = np.delete(cc, 0, 0)
      cc = np.delete(cc, 0, 0)
      np.savetxt(f"test_feature.txt", cc, fmt="%.6f")
      print(cc)
      print(cc.shape)

def test(loader):
     model.eval()
     y_predicted = []
     cc = np.zeros((2, 3))
     correct = 0
     predTime = time.time()
     for data in loader:  # Iterate in batches over the training/test dataset.
         out = model(data.x, data.edge_index, data.batch)
         pred = out.argmax(dim=1)  # Use the class with highest probability.
         y_predicted.extend(pred.tolist())
         correct += int((pred == data.y).sum())  # Check against ground-truth labels.
     print(time.time()-predTime)
     return correct / len(loader.dataset), y_predicted # Derive ratio of correct predictions.

flag = 0
acc = []
pre = []
rec = []
f1 = []
# start_time = time.time()
for epoch in range(1, 31):
    train(epoch)
    # train_acc = test(train_loader)
    test_acc, y_predicted = test(test_loader)
    # if epoch % 10 == 0 and epoch != 0:
    #   print(f'Epoch {epoch}:', (time.time() - start_time))
    #   print(f'Epoch: {epoch:03d}, Train Acc: , Test Acc: {test_acc:.4f}')
    #   print(classification_report(y_test, y_predicted, digits=4))
    #   print(accuracy_score(y_test, y_predicted))
    #   cm = confusion_matrix(y_test, y_predicted)
    #   cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ["Arithmetic", "Clean", "Reentrancy"])

    #   cm_display.plot()
    #   plt.show()
    #   print('=================================================================')
    #   start_time = time.time()
    if epoch == 10:
      flag = 1
    if flag == 1:
      print(epoch)
      metric = [i for i in classification_report(y_test, y_predicted, digits=4).split('\n')[-2].split(' ')[2:-1] if i != '']
      acc.append(round(accuracy_score(y_test, y_predicted), 4))
      pre.append(metric[0])
      rec.append(metric[1])
      f1.append(metric[2])
```
